### Slide 16 Outlier - Deteksi

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Ambil hanya kolom yang akan dicek outliersnya
outlier_cols = [
    'usia',
    'jumlah_tanggunga',
    'pendapatan_per_b',
    'lama_bekerja_tah',
    'jumlah_kredit_ak',
    'total_kredit',
    'sisa_bayar_kredi',
    'tenor_bulan',
    'bunga_per_tahun',
    'cicilan_per_bula',
    'rasio_utang_pend',
    'jumlah_tunggakan',
    'skor_kredit_sebe'
]

# =========================
# 1. DETEKSI OUTLIER FORMAL
# =========================

outlier_summary = []

for col in outlier_cols:
    data = df[col].dropna()
    
    # IQR Method
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_iqr = Q1 - 1.5 * IQR
    upper_iqr = Q3 + 1.5 * IQR
    
    outlier_iqr = ((df[col] < lower_iqr) | (df[col] > upper_iqr)).sum()
    
    # Z-Score Method
    mean = data.mean()
    std = data.std()
    
    if std != 0:
        z_score = (df[col] - mean) / std
        outlier_z = (np.abs(z_score) > 3).sum()
    else:
        outlier_z = 0
    
    outlier_summary.append({
        'kolom': col,
        'jumlah_outlier_IQR': outlier_iqr,
        'persen_outlier_IQR (%)': round(outlier_iqr / len(df) * 100, 2),
        'jumlah_outlier_Zscore': outlier_z,
        'persen_outlier_Zscore (%)': round(outlier_z / len(df) * 100, 2),
        'batas_bawah_IQR': lower_iqr,
        'batas_atas_IQR': upper_iqr
    })

outlier_summary = pd.DataFrame(outlier_summary)
outlier_summary = outlier_summary.sort_values(
    by='persen_outlier_IQR (%)', 
    ascending=False
)

display(outlier_summary)


# =========================
# 2. DETEKSI NON-FORMAL GRAFIK
# =========================

for col in outlier_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    # Boxplot
    axes[0].boxplot(df[col].dropna(), vert=False)
    axes[0].set_title(f'Boxplot - {col}')
    axes[0].set_xlabel(col)
    
    # Histogram
    axes[1].hist(df[col].dropna(), bins=30)
    axes[1].set_title(f'Histogram - {col}')
    axes[1].set_xlabel(col)
    axes[1].set_ylabel('Frekuensi')
    
    plt.tight_layout()
    plt.show()
```

### Slide 18 EDA, SFA, & Korelasi

```python
# ============================================================
# SLIDE 18 - EDA & SINGLE FACTOR ANALYSIS
# Korelasi, Mann-Whitney, Chi-square, Cramer's V, IV, Gini/AR
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import mannwhitneyu, chi2_contingency, spearmanr
from sklearn.metrics import roc_auc_score

target = 'is_default'

# ------------------------------------------------------------
# 1. Tentukan variabel kandidat
# ------------------------------------------------------------

numeric_features = [
    'usia',
    'jumlah_tanggunga',
    'pendapatan_per_b',
    'lama_bekerja_tah',
    'jumlah_kredit_ak',
    'total_kredit',
    'sisa_bayar_kredi',
    'tenor_bulan',
    'bunga_per_tahun',
    'cicilan_per_bula',
    'rasio_utang_pend',
    'jumlah_tunggakan',
    'skor_kredit_sebe'
]

categorical_features = [
    'jenis_kelamin',
    'status_pernikaha',
    'jenis_pekerjaan',
    'pendidikan',
    'kota',
    'jenis_kredit',
    'riwayat_tunggaka'
]

# Ambil hanya kolom yang memang ada di df
numeric_features = [col for col in numeric_features if col in df.columns]
categorical_features = [col for col in categorical_features if col in df.columns]


# ------------------------------------------------------------
# 2. Function IV dan WOE
# ------------------------------------------------------------

def calculate_iv(data, feature, target, bins=5):
    temp = data[[feature, target]].copy()
    
    if pd.api.types.is_numeric_dtype(temp[feature]):
        try:
            temp['bin'] = pd.qcut(temp[feature], q=bins, duplicates='drop')
        except:
            temp['bin'] = temp[feature]
    else:
        temp['bin'] = temp[feature].astype('object').where(temp[feature].notna(), 'Missing')
    
    grouped = temp.groupby('bin', dropna=False)[target].agg(['count', 'sum'])
    grouped = grouped.rename(columns={'sum': 'bad'})
    grouped['good'] = grouped['count'] - grouped['bad']
    
    total_good = grouped['good'].sum()
    total_bad = grouped['bad'].sum()
    
    # Smoothing agar tidak error jika ada bin dengan good/bad = 0
    grouped['dist_good'] = (grouped['good'] + 0.5) / (total_good + 0.5 * len(grouped))
    grouped['dist_bad'] = (grouped['bad'] + 0.5) / (total_bad + 0.5 * len(grouped))
    
    grouped['woe'] = np.log(grouped['dist_good'] / grouped['dist_bad'])
    grouped['iv_component'] = (grouped['dist_good'] - grouped['dist_bad']) * grouped['woe']
    
    iv = grouped['iv_component'].sum()
    return iv


def interpret_iv(iv):
    if pd.isna(iv):
        return 'Tidak tersedia'
    elif iv < 0.02:
        return 'Sangat lemah'
    elif iv < 0.10:
        return 'Lemah'
    elif iv < 0.30:
        return 'Menengah'
    elif iv < 0.50:
        return 'Kuat'
    else:
        return 'Sangat kuat'


# ------------------------------------------------------------
# 3. Function Cramer's V untuk kategorik
# ------------------------------------------------------------

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2, p, dof, expected = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    
    if min(r - 1, k - 1) == 0:
        return np.nan
    
    return np.sqrt((chi2 / n) / min(r - 1, k - 1))


# ------------------------------------------------------------
# 4. Screening variabel numerik
# ------------------------------------------------------------

results = []

for col in numeric_features:
    temp = df[[col, target]].dropna()
    
    good = temp[temp[target] == 0][col]
    bad = temp[temp[target] == 1][col]
    
    # Mann-Whitney U Test
    try:
        stat, p_value = mannwhitneyu(good, bad, alternative='two-sided')
    except:
        p_value = np.nan
    
    # Spearman correlation terhadap target
    try:
        spearman_corr, spearman_p = spearmanr(temp[col], temp[target])
    except:
        spearman_corr, spearman_p = np.nan, np.nan
    
    # Gini / AR univariat
    try:
        auc = roc_auc_score(temp[target], temp[col])
        gini_raw = 2 * auc - 1
        gini_abs = abs(gini_raw)
    except:
        auc, gini_raw, gini_abs = np.nan, np.nan, np.nan
    
    # IV
    try:
        iv = calculate_iv(df, col, target, bins=5)
    except:
        iv = np.nan
    
    results.append({
        'variabel': col,
        'jenis_variabel': 'Numerik',
        'uji_signifikansi': 'Mann-Whitney U',
        'p_value': p_value,
        'spearman_corr': spearman_corr,
        'iv': iv,
        'gini_ar_univariat': gini_abs,
        'cramers_v': np.nan
    })


# ------------------------------------------------------------
# 5. Screening variabel kategorik
# ------------------------------------------------------------

for col in categorical_features:
    temp = df[[col, target]].dropna()
    
    # Chi-square
    try:
        contingency = pd.crosstab(temp[col], temp[target])
        chi2, p_value, dof, expected = chi2_contingency(contingency)
    except:
        p_value = np.nan
    
    # Cramer's V
    try:
        cv = cramers_v(temp[col], temp[target])
    except:
        cv = np.nan
    
    # IV
    try:
        iv = calculate_iv(df, col, target)
    except:
        iv = np.nan
    
    results.append({
        'variabel': col,
        'jenis_variabel': 'Kategorik',
        'uji_signifikansi': 'Chi-square',
        'p_value': p_value,
        'spearman_corr': np.nan,
        'iv': iv,
        'gini_ar_univariat': np.nan,
        'cramers_v': cv
    })


# ------------------------------------------------------------
# 6. Tabel hasil Slide 18
# ------------------------------------------------------------

slide18_screening = pd.DataFrame(results)

slide18_screening['interpretasi_iv'] = slide18_screening['iv'].apply(interpret_iv)

slide18_screening['signifikan_5persen'] = np.where(
    slide18_screening['p_value'] < 0.05,
    'Ya',
    'Tidak'
)

# Skor screening awal
def screening_decision(row):
    score = 0
    
    if pd.notna(row['p_value']) and row['p_value'] < 0.05:
        score += 1
    
    if pd.notna(row['iv']) and row['iv'] >= 0.02:
        score += 1
    
    if row['jenis_variabel'] == 'Numerik':
        if pd.notna(row['gini_ar_univariat']) and row['gini_ar_univariat'] >= 0.05:
            score += 1
    
    if row['jenis_variabel'] == 'Kategorik':
        if pd.notna(row['cramers_v']) and row['cramers_v'] >= 0.10:
            score += 1
    
    if score >= 2:
        return 'Dipertimbangkan'
    elif score == 1:
        return 'Dipertimbangkan dengan catatan'
    else:
        return 'Tidak diprioritaskan'

slide18_screening['keputusan_screening'] = slide18_screening.apply(screening_decision, axis=1)

slide18_screening = slide18_screening.sort_values(
    by=['keputusan_screening', 'iv', 'gini_ar_univariat'],
    ascending=[True, False, False]
)

display(slide18_screening)


# ------------------------------------------------------------
# 7. Korelasi antar variabel numerik
# ------------------------------------------------------------

corr_cols = numeric_features + [target]
corr_cols = [col for col in corr_cols if col in df.columns]

corr_matrix = df[corr_cols].corr(method='spearman')

# Korelasi terhadap target
corr_target = corr_matrix[[target]].drop(target).sort_values(by=target, ascending=False)
display(corr_target)

# Pasangan korelasi tinggi antar prediktor numerik
high_corr = []

for i in range(len(numeric_features)):
    for j in range(i + 1, len(numeric_features)):
        var1 = numeric_features[i]
        var2 = numeric_features[j]
        corr_value = corr_matrix.loc[var1, var2]
        
        if abs(corr_value) >= 0.70:
            high_corr.append({
                'variabel_1': var1,
                'variabel_2': var2,
                'korelasi_spearman': corr_value,
                'catatan': 'Potensi multikolinearitas'
            })

high_corr_df = pd.DataFrame(high_corr)
display(high_corr_df)


# ------------------------------------------------------------
# 8. Heatmap korelasi untuk Slide 18
# ------------------------------------------------------------

plt.figure(figsize=(12, 8))
plt.imshow(corr_matrix, aspect='auto')
plt.colorbar(label='Spearman Correlation')
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title('Heatmap Korelasi Spearman Variabel Numerik')
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# 9. Tabel ringkas untuk ditampilkan di Slide 18
# ------------------------------------------------------------

slide18_output = slide18_screening[
    [
        'variabel',
        'jenis_variabel',
        'uji_signifikansi',
        'p_value',
        'signifikan_5persen',
        'iv',
        'interpretasi_iv',
        'gini_ar_univariat',
        'cramers_v',
        'keputusan_screening'
    ]
].copy()

display(slide18_output)
```

### Slide 19 Feature Selection

```python
    # ============================================================
    # SLIDE 19 - FEATURE SELECTION FINAL
    # WOE Transform, Logistic Regression, p-value, VIF, Gini/AR, KS
    # ============================================================

    import numpy as np
    import pandas as pd
    import statsmodels.api as sm

    from sklearn.model_selection import train_test_split
    from sklearn.metrics import roc_auc_score
    from statsmodels.stats.outliers_influence import variance_inflation_factor

    target = 'is_default'


    # ------------------------------------------------------------
    # 1. Ambil kandidat dari Slide 18
    # ------------------------------------------------------------

    candidate_vars = slide18_screening[
        slide18_screening['keputusan_screening'] != 'Tidak diprioritaskan'
    ]['variabel'].tolist()

    # Buang kolom yang tidak boleh masuk model
    blocked_vars = [
        'id_nasabah',
        'nama',
        target
    ]

    candidate_vars = [col for col in candidate_vars if col not in blocked_vars]

    print("Jumlah kandidat awal:", len(candidate_vars))
    print(candidate_vars)


    # ------------------------------------------------------------
    # 2. Function WOE table dan transform
    # ------------------------------------------------------------

    def make_woe_table_for_transform(data, feature, target, bins=5):
        temp = data[[feature, target]].copy()
        
        is_numeric = pd.api.types.is_numeric_dtype(temp[feature])
        
        if is_numeric:
            try:
                temp['bin'] = pd.qcut(temp[feature], q=bins, duplicates='drop')
            except:
                temp['bin'] = temp[feature]
        else:
            temp['bin'] = temp[feature].astype('object').where(temp[feature].notna(), 'Missing')
        
        grouped = temp.groupby('bin', dropna=False)[target].agg(['count', 'sum'])
        grouped = grouped.rename(columns={'sum': 'bad'})
        grouped['good'] = grouped['count'] - grouped['bad']
        
        total_good = grouped['good'].sum()
        total_bad = grouped['bad'].sum()
        
        grouped['dist_good'] = (grouped['good'] + 0.5) / (total_good + 0.5 * len(grouped))
        grouped['dist_bad'] = (grouped['bad'] + 0.5) / (total_bad + 0.5 * len(grouped))
        
        grouped['woe'] = np.log(grouped['dist_good'] / grouped['dist_bad'])
        grouped['iv_component'] = (grouped['dist_good'] - grouped['dist_bad']) * grouped['woe']
        grouped['bad_rate'] = grouped['bad'] / grouped['count']
        
        iv = grouped['iv_component'].sum()
        
        return grouped.reset_index(), iv, is_numeric


    def apply_woe_transform(data, feature, woe_table, is_numeric, bins=5):
        out = pd.Series(index=data.index, dtype='float64')
        
        if is_numeric:
            try:
                binned = pd.qcut(data[feature], q=bins, duplicates='drop')
            except:
                binned = data[feature]
            
            woe_map = woe_table.set_index('bin')['woe'].to_dict()
            out = binned.map(woe_map).astype(float)
            
        else:
            temp = data[feature].astype('object').where(data[feature].notna(), 'Missing')
            woe_map = woe_table.set_index('bin')['woe'].to_dict()
            out = temp.map(woe_map).astype(float)
        
        return out


    # ------------------------------------------------------------
    # 3. Transform semua kandidat menjadi WOE
    # ------------------------------------------------------------

    df_model = df.copy()

    woe_info = {}
    iv_final_rows = []

    for var in candidate_vars:
        try:
            woe_table, iv, is_numeric = make_woe_table_for_transform(df_model, var, target, bins=5)
            woe_col = 'WOE_' + var
            
            df_model[woe_col] = apply_woe_transform(
                df_model,
                var,
                woe_table,
                is_numeric,
                bins=5
            )
            
            woe_info[var] = {
                'woe_table': woe_table,
                'iv': iv,
                'is_numeric': is_numeric,
                'woe_col': woe_col
            }
            
            iv_final_rows.append({
                'variabel': var,
                'woe_col': woe_col,
                'iv': iv
            })
        
        except Exception as e:
            print(f"Gagal WOE untuk {var}: {e}")

    iv_final_df = pd.DataFrame(iv_final_rows).sort_values(by='iv', ascending=False)
    display(iv_final_df)

    woe_cols = iv_final_df['woe_col'].tolist()

    # Drop missing hasil WOE
    df_model = df_model.dropna(subset=woe_cols + [target]).copy()

    X = df_model[woe_cols]
    y = df_model[target].astype(int)


    # ------------------------------------------------------------
    # 4. Train-test split
    # ------------------------------------------------------------

    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=0.30,
        random_state=42,
        stratify=y
    )


    # ------------------------------------------------------------
    # 5. Function VIF
    # ------------------------------------------------------------

    def calculate_vif_table(X):
        X_const = sm.add_constant(X)
        rows = []
        
        for i, col in enumerate(X_const.columns):
            if col == 'const':
                continue
            
            try:
                vif_value = variance_inflation_factor(X_const.values, i)
            except:
                vif_value = np.nan
            
            rows.append({
                'woe_col': col,
                'VIF': vif_value
            })
        
        return pd.DataFrame(rows).sort_values(by='VIF', ascending=False)


    # ------------------------------------------------------------
    # 6. Function KS
    # ------------------------------------------------------------

    def ks_statistic(y_true, y_score):
        data = pd.DataFrame({
            'y': y_true,
            'score': y_score
        }).sort_values('score', ascending=False)
        
        total_bad = data['y'].sum()
        total_good = len(data) - total_bad
        
        data['cum_bad'] = data['y'].cumsum() / total_bad
        data['cum_good'] = (1 - data['y']).cumsum() / total_good
        data['ks'] = abs(data['cum_bad'] - data['cum_good'])
        
        return data['ks'].max()


    # ------------------------------------------------------------
    # 7. Logistic Regression awal
    # ------------------------------------------------------------

    X_train_sm = sm.add_constant(X_train)
    X_test_sm = sm.add_constant(X_test)

    logit_model = sm.Logit(y_train, X_train_sm).fit(maxiter=200)

    print(logit_model.summary())

    pvalue_table = pd.DataFrame({
        'woe_col': logit_model.pvalues.index,
        'coef': logit_model.params.values,
        'p_value_model': logit_model.pvalues.values
    })

    pvalue_table = pvalue_table[pvalue_table['woe_col'] != 'const'].copy()

    display(pvalue_table.sort_values(by='p_value_model'))


    # ------------------------------------------------------------
    # 8. VIF awal
    # ------------------------------------------------------------

    vif_table = calculate_vif_table(X_train)
    display(vif_table)


    # ------------------------------------------------------------
    # 9. Performance model awal: Gini/AR dan KS
    # ------------------------------------------------------------

    pd_train = logit_model.predict(X_train_sm)
    pd_test = logit_model.predict(X_test_sm)

    auc_train = roc_auc_score(y_train, pd_train)
    auc_test = roc_auc_score(y_test, pd_test)

    gini_train = 2 * auc_train - 1
    gini_test = 2 * auc_test - 1

    ks_train = ks_statistic(y_train, pd_train)
    ks_test = ks_statistic(y_test, pd_test)

    model_perf = pd.DataFrame({
        'dataset': ['Train', 'Test'],
        'AUC': [auc_train, auc_test],
        'Gini_AR': [gini_train, gini_test],
        'KS': [ks_train, ks_test]
    })

    display(model_perf)


    # ------------------------------------------------------------
    # 10. Compile hasil semua uji
    # ------------------------------------------------------------

    # Mapping WOE_col ke nama variabel asli
    mapping_var = iv_final_df[['variabel', 'woe_col', 'iv']].copy()

    slide19_compile = mapping_var.merge(
        pvalue_table[['woe_col', 'coef', 'p_value_model']],
        on='woe_col',
        how='left'
    ).merge(
        vif_table,
        on='woe_col',
        how='left'
    )

    # Gabungkan hasil Slide 18
    slide19_compile = slide19_compile.merge(
        slide18_screening[
            [
                'variabel',
                'jenis_variabel',
                'p_value',
                'signifikan_5persen',
                'spearman_corr',
                'gini_ar_univariat',
                'cramers_v',
                'interpretasi_iv',
                'keputusan_screening'
            ]
        ],
        on='variabel',
        how='left'
    )

    # Keputusan berdasarkan seluruh uji
    def final_decision(row):
        alasan = []
        
        # 1. Signifikan di uji awal
        if pd.notna(row['p_value']) and row['p_value'] < 0.05:
            alasan.append('signifikan di uji awal')
        
        # 2. IV
        if pd.notna(row['iv']) and row['iv'] >= 0.02:
            alasan.append('IV memadai')
        
        # 3. Gini/AR univariat
        if pd.notna(row['gini_ar_univariat']) and row['gini_ar_univariat'] >= 0.05:
            alasan.append('Gini univariat memadai')
        
        # 4. Cramer's V untuk kategorik
        if pd.notna(row['cramers_v']) and row['cramers_v'] >= 0.10:
            alasan.append('asosiasi kategorik cukup')
        
        # 5. P-value model logistic
        if pd.notna(row['p_value_model']) and row['p_value_model'] < 0.05:
            alasan.append('signifikan di model')
        
        # 6. VIF aman
        if pd.notna(row['VIF']) and row['VIF'] < 10:
            alasan.append('VIF aman')
        
        # Decision rules
        if pd.notna(row['VIF']) and row['VIF'] >= 10:
            return 'Tidak dipakai', 'VIF tinggi / indikasi multikolinearitas'
        
        if pd.notna(row['p_value_model']) and row['p_value_model'] >= 0.05:
            if row['iv'] >= 0.10:
                return 'Dipertimbangkan', 'Tidak signifikan di model, tetapi IV cukup kuat; perlu business judgement'
            else:
                return 'Tidak dipakai', 'Tidak signifikan di model dan IV kurang kuat'
        
        if len(alasan) >= 4:
            return 'Dipakai', ', '.join(alasan)
        elif len(alasan) >= 2:
            return 'Dipertimbangkan', ', '.join(alasan)
        else:
            return 'Tidak dipakai', ', '.join(alasan) if alasan else 'Tidak cukup bukti statistik'

    decision_result = slide19_compile.apply(final_decision, axis=1)

    slide19_compile['keputusan_final'] = [x[0] for x in decision_result]
    slide19_compile['alasan_final'] = [x[1] for x in decision_result]

    slide19_compile = slide19_compile.sort_values(
        by=['keputusan_final', 'iv', 'gini_ar_univariat'],
        ascending=[True, False, False]
    )

    display(slide19_compile)


    # ------------------------------------------------------------
    # 11. Tabel ringkas untuk ditampilkan di Slide 19
    # ------------------------------------------------------------

    slide19_output = slide19_compile[
        [
            'variabel',
            'jenis_variabel',
            'p_value',
            'iv',
            'interpretasi_iv',
            'gini_ar_univariat',
            'cramers_v',
            'p_value_model',
            'VIF',
            'keputusan_final',
            'alasan_final'
        ]
    ].copy()

    display(slide19_output)


    # ------------------------------------------------------------
    # 12. List variabel final yang dipakai
    # ------------------------------------------------------------

    final_features = slide19_compile[
        slide19_compile['keputusan_final'].isin(['Dipakai', 'Dipertimbangkan'])
    ]['variabel'].tolist()

    final_woe_cols = ['WOE_' + col for col in final_features if 'WOE_' + col in df_model.columns]

    print("Variabel final yang dipakai/dipertimbangkan:")
    print(final_features)

    print("\nKolom WOE final:")
    print(final_woe_cols)
```

### Slide 21 Modelling Logistic Regression

```python
# ============================================================
# SLIDE 21 - MODEL 1: LOGISTIC REGRESSION
# ============================================================

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Pisahkan jenis kolom
numeric_features_logreg = [
    'pendapatan_per_b',
    'jumlah_tunggakan',
    'rasio_utang_pend',
    'usia',
    'skor_kredit_sebe'
]

categorical_features_logreg = [
    'jenis_pekerjaan'
]

# Preprocessing
preprocessor_logreg = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features_logreg),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features_logreg)
    ]
)

# Model Logistic Regression
logreg_model = LogisticRegression(
    class_weight='balanced',
    max_iter=1000,
    random_state=42
)

# Pipeline
logreg_pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor_logreg),
        ('model', logreg_model)
    ]
)

# Train model
logreg_pipeline.fit(X_train, y_train)

print("Model Logistic Regression berhasil dilatih.")
print("Fitur yang digunakan:")
print(final_features_logreg)
print("\nClass weight yang digunakan: balanced")
```

```python
# ============================================================
# SLIDE 22 - INTERPRETASI LOGISTIC REGRESSION
# ============================================================

import numpy as np
import pandas as pd

# Ambil nama fitur setelah preprocessing
encoded_feature_names = logreg_pipeline.named_steps['preprocessor'].get_feature_names_out()

# Ambil koefisien model
coef_values = logreg_pipeline.named_steps['model'].coef_[0]

# Buat tabel koefisien
coef_table = pd.DataFrame({
    'feature': encoded_feature_names,
    'coefficient': coef_values
})

# Odds ratio
coef_table['odds_ratio'] = np.exp(coef_table['coefficient'])

# Interpretasi arah
coef_table['arah_pengaruh'] = np.where(
    coef_table['coefficient'] > 0,
    'Meningkatkan risiko default',
    'Menurunkan risiko default'
)

# Besarnya pengaruh absolut
coef_table['abs_coefficient'] = coef_table['coefficient'].abs()

# Urutkan berdasarkan pengaruh terbesar
coef_table = coef_table.sort_values(by='abs_coefficient', ascending=False)

display(coef_table)
```

```python
# ============================================================
# EVALUASI MODEL LOGISTIC REGRESSION
# Metrics: Confusion Matrix, Accuracy, Precision, Recall,
# F1, Specificity, ROC-AUC, PR-AUC, KS, Log Loss, Brier Score,
# Balanced Accuracy, Classification Report, ROC Curve,
# Precision-Recall Curve, Threshold Analysis
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score,
    log_loss,
    brier_score_loss,
    balanced_accuracy_score,
    matthews_corrcoef,
    cohen_kappa_score,
    roc_curve,
    precision_recall_curve
)

# ============================================================
# 1. Prediksi pada data train dan test
# ============================================================

# Prediksi kelas default/non-default
y_train_pred = logreg_pipeline.predict(X_train)
y_test_pred = logreg_pipeline.predict(X_test)

# Prediksi probabilitas default
y_train_proba = logreg_pipeline.predict_proba(X_train)[:, 1]
y_test_proba = logreg_pipeline.predict_proba(X_test)[:, 1]


# ============================================================
# 2. Function evaluasi lengkap
# ============================================================

def evaluate_model(y_true, y_pred, y_proba, dataset_name='Data'):
    cm = confusion_matrix(y_true, y_pred)
    
    tn, fp, fn, tp = cm.ravel()
    
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    specificity = tn / (tn + fp) if (tn + fp) != 0 else np.nan
    npv = tn / (tn + fn) if (tn + fn) != 0 else np.nan
    
    fpr = fp / (fp + tn) if (fp + tn) != 0 else np.nan
    fnr = fn / (fn + tp) if (fn + tp) != 0 else np.nan
    
    balanced_acc = balanced_accuracy_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_proba)
    pr_auc = average_precision_score(y_true, y_proba)
    logloss = log_loss(y_true, y_proba)
    brier = brier_score_loss(y_true, y_proba)
    mcc = matthews_corrcoef(y_true, y_pred)
    kappa = cohen_kappa_score(y_true, y_pred)
    
    return {
        'dataset': dataset_name,
        'TN': tn,
        'FP': fp,
        'FN': fn,
        'TP': tp,
        'accuracy': accuracy,
        'balanced_accuracy': balanced_acc,
        'precision': precision,
        'recall_sensitivity': recall,
        'specificity': specificity,
        'f1_score': f1,
        'NPV': npv,
        'FPR': fpr,
        'FNR': fnr,
        'ROC_AUC': roc_auc,
        'PR_AUC': pr_auc,
        'log_loss': logloss,
        'brier_score': brier,
        'MCC': mcc,
        'cohen_kappa': kappa
    }


# ============================================================
# 3. Buat tabel evaluasi train dan test
# ============================================================

eval_train = evaluate_model(y_train, y_train_pred, y_train_proba, 'Train')
eval_test = evaluate_model(y_test, y_test_pred, y_test_proba, 'Test')

eval_table = pd.DataFrame([eval_train, eval_test])

# Format angka agar rapi
eval_table_rounded = eval_table.copy()
metric_cols = eval_table_rounded.columns.drop(['dataset', 'TN', 'FP', 'FN', 'TP'])

eval_table_rounded[metric_cols] = eval_table_rounded[metric_cols].round(4)

display(eval_table_rounded)


# ============================================================
# 4. Classification report
# ============================================================

print("CLASSIFICATION REPORT - TRAIN")
print(classification_report(y_train, y_train_pred, target_names=['Non-Default', 'Default']))

print("\nCLASSIFICATION REPORT - TEST")
print(classification_report(y_test, y_test_pred, target_names=['Non-Default', 'Default']))


# ============================================================
# 5. Confusion matrix train dan test
# ============================================================

cm_train = confusion_matrix(y_train, y_train_pred)
cm_test = confusion_matrix(y_test, y_test_pred)

cm_train_df = pd.DataFrame(
    cm_train,
    index=['Actual Non-Default', 'Actual Default'],
    columns=['Pred Non-Default', 'Pred Default']
)

cm_test_df = pd.DataFrame(
    cm_test,
    index=['Actual Non-Default', 'Actual Default'],
    columns=['Pred Non-Default', 'Pred Default']
)

print("CONFUSION MATRIX - TRAIN")
display(cm_train_df)

print("CONFUSION MATRIX - TEST")
display(cm_test_df)


# ============================================================
# 6. ROC Curve
# ============================================================

fpr_train, tpr_train, thresholds_train = roc_curve(y_train, y_train_proba)
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, y_test_proba)

plt.figure(figsize=(7, 5))
plt.plot(fpr_train, tpr_train, label=f'Train ROC-AUC = {roc_auc_score(y_train, y_train_proba):.3f}')
plt.plot(fpr_test, tpr_test, label=f'Test ROC-AUC = {roc_auc_score(y_test, y_test_proba):.3f}')
plt.plot([0, 1], [0, 1], linestyle='--', label='Random Model')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate / Recall')
plt.title('ROC Curve - Logistic Regression')
plt.legend()
plt.grid(True)
plt.show()


# ============================================================
# 7. Precision-Recall Curve
# ============================================================

precision_train, recall_train, threshold_pr_train = precision_recall_curve(y_train, y_train_proba)
precision_test, recall_test, threshold_pr_test = precision_recall_curve(y_test, y_test_proba)

plt.figure(figsize=(7, 5))
plt.plot(recall_train, precision_train, label=f'Train PR-AUC = {average_precision_score(y_train, y_train_proba):.3f}')
plt.plot(recall_test, precision_test, label=f'Test PR-AUC = {average_precision_score(y_test, y_test_proba):.3f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve - Logistic Regression')
plt.legend()
plt.grid(True)
plt.show()


# ============================================================
# 8. KS Statistic
# ============================================================

def ks_table(y_true, y_proba, n_bins=10):
    data = pd.DataFrame({
        'actual': y_true,
        'proba_default': y_proba
    })
    
    data['bucket'] = pd.qcut(
        data['proba_default'],
        q=n_bins,
        duplicates='drop'
    )
    
    grouped = data.groupby('bucket', observed=False).agg(
        total=('actual', 'count'),
        bad=('actual', 'sum')
    ).reset_index()
    
    grouped['good'] = grouped['total'] - grouped['bad']
    
    # Urutkan dari probability default tertinggi
    grouped = grouped.sort_values(by='bucket', ascending=False).reset_index(drop=True)
    
    grouped['bad_rate'] = grouped['bad'] / grouped['total']
    grouped['dist_bad'] = grouped['bad'] / grouped['bad'].sum()
    grouped['dist_good'] = grouped['good'] / grouped['good'].sum()
    grouped['cum_bad'] = grouped['dist_bad'].cumsum()
    grouped['cum_good'] = grouped['dist_good'].cumsum()
    grouped['ks'] = abs(grouped['cum_bad'] - grouped['cum_good'])
    
    return grouped

ks_train_table = ks_table(y_train, y_train_proba)
ks_test_table = ks_table(y_test, y_test_proba)

ks_train_value = ks_train_table['ks'].max()
ks_test_value = ks_test_table['ks'].max()

print(f"KS Train: {ks_train_value:.4f}")
print(f"KS Test : {ks_test_value:.4f}")

print("\nKS TABLE - TEST")
display(ks_test_table)


# ============================================================
# 9. Gini / Accuracy Ratio
# ============================================================

gini_train = 2 * roc_auc_score(y_train, y_train_proba) - 1
gini_test = 2 * roc_auc_score(y_test, y_test_proba) - 1

gini_table = pd.DataFrame({
    'dataset': ['Train', 'Test'],
    'AUC': [
        roc_auc_score(y_train, y_train_proba),
        roc_auc_score(y_test, y_test_proba)
    ],
    'Gini_AR': [gini_train, gini_test],
    'KS': [ks_train_value, ks_test_value]
})

gini_table[['AUC', 'Gini_AR', 'KS']] = gini_table[['AUC', 'Gini_AR', 'KS']].round(4)

display(gini_table)


# ============================================================
# 10. Threshold analysis
# ============================================================

def threshold_analysis(y_true, y_proba, thresholds=np.arange(0.1, 1.0, 0.05)):
    rows = []
    
    for threshold in thresholds:
        y_pred_threshold = (y_proba >= threshold).astype(int)
        
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred_threshold).ravel()
        
        accuracy = accuracy_score(y_true, y_pred_threshold)
        precision = precision_score(y_true, y_pred_threshold, zero_division=0)
        recall = recall_score(y_true, y_pred_threshold, zero_division=0)
        f1 = f1_score(y_true, y_pred_threshold, zero_division=0)
        specificity = tn / (tn + fp) if (tn + fp) != 0 else np.nan
        
        rows.append({
            'threshold': threshold,
            'TP': tp,
            'FP': fp,
            'TN': tn,
            'FN': fn,
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'specificity': specificity,
            'f1_score': f1
        })
    
    return pd.DataFrame(rows)

threshold_df = threshold_analysis(y_test, y_test_proba)
threshold_df_rounded = threshold_df.copy()
threshold_df_rounded[['accuracy', 'precision', 'recall', 'specificity', 'f1_score']] = (
    threshold_df_rounded[['accuracy', 'precision', 'recall', 'specificity', 'f1_score']]
    .round(4)
)

display(threshold_df_rounded)

# Threshold terbaik berdasarkan F1-score
best_threshold_row = threshold_df.loc[threshold_df['f1_score'].idxmax()]
best_threshold = best_threshold_row['threshold']

print(f"Best threshold berdasarkan F1-score: {best_threshold:.2f}")
display(best_threshold_row.to_frame().T)


# ============================================================
# 11. Lift table / Decile analysis
# ============================================================

def lift_table(y_true, y_proba, n_bins=10):
    data = pd.DataFrame({
        'actual': y_true,
        'proba_default': y_proba
    })
    
    data['decile'] = pd.qcut(
        data['proba_default'],
        q=n_bins,
        labels=False,
        duplicates='drop'
    )
    
    # Decile 1 = risiko tertinggi
    data['decile'] = data['decile'].max() - data['decile'] + 1
    
    grouped = data.groupby('decile').agg(
        total=('actual', 'count'),
        bad=('actual', 'sum'),
        avg_proba=('proba_default', 'mean')
    ).reset_index()
    
    grouped['good'] = grouped['total'] - grouped['bad']
    grouped['bad_rate'] = grouped['bad'] / grouped['total']
    grouped['cum_bad'] = grouped['bad'].cumsum()
    grouped['cum_total'] = grouped['total'].cumsum()
    grouped['cum_bad_capture_rate'] = grouped['cum_bad'] / grouped['bad'].sum()
    grouped['population_pct'] = grouped['cum_total'] / grouped['total'].sum()
    
    baseline_bad_rate = data['actual'].mean()
    grouped['lift'] = grouped['bad_rate'] / baseline_bad_rate
    
    return grouped

lift_test = lift_table(y_test, y_test_proba)
lift_test_rounded = lift_test.copy()

for col in ['avg_proba', 'bad_rate', 'cum_bad_capture_rate', 'population_pct', 'lift']:
    lift_test_rounded[col] = lift_test_rounded[col].round(4)

display(lift_test_rounded)


# ============================================================
# 12. Plot actual default rate per decile
# ============================================================

plt.figure(figsize=(8, 5))
plt.bar(lift_test['decile'].astype(str), lift_test['bad_rate'])
plt.xlabel('Decile Risiko Prediksi')
plt.ylabel('Actual Default Rate')
plt.title('Actual Default Rate per Decile - Test Data')
plt.grid(axis='y')
plt.show()
```

```python
def calculate_iv_interaction(data, feature, target='is_default'):
    temp = data[[feature, target]].copy()
    temp[feature] = temp[feature].astype('object').where(temp[feature].notna(), 'Missing')
    
    grouped = temp.groupby(feature)[target].agg(['count', 'sum'])
    grouped = grouped.rename(columns={'sum': 'bad'})
    grouped['good'] = grouped['count'] - grouped['bad']
    
    total_good = grouped['good'].sum()
    total_bad = grouped['bad'].sum()
    
    # smoothing
    grouped['dist_good'] = (grouped['good'] + 0.5) / (total_good + 0.5 * len(grouped))
    grouped['dist_bad'] = (grouped['bad'] + 0.5) / (total_bad + 0.5 * len(grouped))
    
    grouped['woe'] = np.log(grouped['dist_good'] / grouped['dist_bad'])
    grouped['iv_component'] = (grouped['dist_good'] - grouped['dist_bad']) * grouped['woe']
    grouped['bad_rate'] = grouped['bad'] / grouped['count']
    
    iv = grouped['iv_component'].sum()
    
    return grouped.reset_index(), iv
```

```python
# ============================================================
# COMPARISON LOGISTIC REGRESSION DENGAN BEBERAPA INTERAKSI
# ============================================================

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    balanced_accuracy_score,
    roc_auc_score,
    average_precision_score,
    classification_report
)

target = 'is_default'

# ============================================================
# 1. Buat ulang variabel interaksi
# ============================================================

df_inter = df.copy()

df_inter['pendapatan_bin'] = pd.qcut(
    df_inter['pendapatan_per_b'],
    q=5,
    duplicates='drop'
)

df_inter['dti_bin'] = pd.qcut(
    df_inter['rasio_utang_pend'],
    q=5,
    duplicates='drop'
)

df_inter['skor_bin'] = pd.qcut(
    df_inter['skor_kredit_sebe'],
    q=5,
    duplicates='drop'
)

df_inter['tunggakan_bin'] = pd.qcut(
    df_inter['jumlah_tunggakan'],
    q=5,
    duplicates='drop'
)

df_inter['interaksi_pendapatan_dti'] = (
    df_inter['pendapatan_bin'].astype(str) + ' | ' +
    df_inter['dti_bin'].astype(str)
)

df_inter['interaksi_pekerjaan_dti'] = (
    df_inter['jenis_pekerjaan'].astype(str) + ' | ' +
    df_inter['dti_bin'].astype(str)
)

df_inter['interaksi_skor_tunggakan'] = (
    df_inter['skor_bin'].astype(str) + ' | ' +
    df_inter['tunggakan_bin'].astype(str)
)


# ============================================================
# 2. Base features
# ============================================================

base_features = [
    'jenis_pekerjaan',
    'pendapatan_per_b',
    'rasio_utang_pend',
    'jumlah_tunggakan',
    'usia',
    'skor_kredit_sebe'
]

numeric_base = [
    'pendapatan_per_b',
    'rasio_utang_pend',
    'jumlah_tunggakan',
    'usia',
    'skor_kredit_sebe'
]

categorical_base = [
    'jenis_pekerjaan'
]


# ============================================================
# 3. Function training dan evaluasi
# ============================================================

def train_eval_logreg_with_features(df_data, feature_list, numeric_cols, categorical_cols, model_name):
    df_model = df_data[feature_list + [target]].dropna().copy()
    
    X = df_model[feature_list]
    y = df_model[target].astype(int)
    
    X_train_i, X_test_i, y_train_i, y_test_i = train_test_split(
        X,
        y,
        test_size=0.20,
        random_state=42,
        stratify=y
    )
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numeric_cols),
            ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_cols)
        ]
    )
    
    model = LogisticRegression(
        class_weight='balanced',
        max_iter=1000,
        random_state=42
    )
    
    pipeline = Pipeline(
        steps=[
            ('preprocessor', preprocessor),
            ('model', model)
        ]
    )
    
    pipeline.fit(X_train_i, y_train_i)
    
    y_pred = pipeline.predict(X_test_i)
    y_proba = pipeline.predict_proba(X_test_i)[:, 1]
    
    result = {
        'model': model_name,
        'accuracy': accuracy_score(y_test_i, y_pred),
        'balanced_accuracy': balanced_accuracy_score(y_test_i, y_pred),
        'precision_default': precision_score(y_test_i, y_pred, zero_division=0),
        'recall_default': recall_score(y_test_i, y_pred, zero_division=0),
        'f1_default': f1_score(y_test_i, y_pred, zero_division=0),
        'ROC_AUC': roc_auc_score(y_test_i, y_proba),
        'PR_AUC': average_precision_score(y_test_i, y_proba)
    }
    
    return result, pipeline, X_test_i, y_test_i, y_pred, y_proba


# ============================================================
# 4. Jalankan beberapa skenario
# ============================================================

experiments = []

# Model A: baseline
res_base, pipe_base, X_test_base, y_test_base, y_pred_base, y_proba_base = train_eval_logreg_with_features(
    df_inter,
    base_features,
    numeric_base,
    categorical_base,
    'Baseline'
)
experiments.append(res_base)


# Model B: + interaksi pendapatan DTI
features_b = base_features + ['interaksi_pendapatan_dti']
numeric_b = numeric_base
categorical_b = categorical_base + ['interaksi_pendapatan_dti']

res_b, pipe_b, X_test_b, y_test_b, y_pred_b, y_proba_b = train_eval_logreg_with_features(
    df_inter,
    features_b,
    numeric_b,
    categorical_b,
    '+ Interaksi Pendapatan x DTI'
)
experiments.append(res_b)


# Model C: + interaksi pekerjaan DTI
features_c = base_features + ['interaksi_pekerjaan_dti']
numeric_c = numeric_base
categorical_c = categorical_base + ['interaksi_pekerjaan_dti']

res_c, pipe_c, X_test_c, y_test_c, y_pred_c, y_proba_c = train_eval_logreg_with_features(
    df_inter,
    features_c,
    numeric_c,
    categorical_c,
    '+ Interaksi Pekerjaan x DTI'
)
experiments.append(res_c)


# Model D: + interaksi skor tunggakan
features_d = base_features + ['interaksi_skor_tunggakan']
numeric_d = numeric_base
categorical_d = categorical_base + ['interaksi_skor_tunggakan']

res_d, pipe_d, X_test_d, y_test_d, y_pred_d, y_proba_d = train_eval_logreg_with_features(
    df_inter,
    features_d,
    numeric_d,
    categorical_d,
    '+ Interaksi Skor x Tunggakan'
)
experiments.append(res_d)


# Model E: semua interaksi
features_e = base_features + [
    'interaksi_pendapatan_dti',
    'interaksi_pekerjaan_dti',
    'interaksi_skor_tunggakan'
]

numeric_e = numeric_base

categorical_e = categorical_base + [
    'interaksi_pendapatan_dti',
    'interaksi_pekerjaan_dti',
    'interaksi_skor_tunggakan'
]

res_e, pipe_e, X_test_e, y_test_e, y_pred_e, y_proba_e = train_eval_logreg_with_features(
    df_inter,
    features_e,
    numeric_e,
    categorical_e,
    '+ Semua Interaksi'
)
experiments.append(res_e)


# ============================================================
# 5. Tabel perbandingan
# ============================================================

interaction_comparison = pd.DataFrame(experiments)

metric_cols = interaction_comparison.columns.drop('model')
interaction_comparison[metric_cols] = interaction_comparison[metric_cols].round(4)

interaction_comparison = interaction_comparison.sort_values(
    by='f1_default',
    ascending=False
)

display(interaction_comparison)
```

```python
# ============================================================
# OPTUNA TUNING - RANDOM FOREST
# Fokus: Recall default, tetapi tetap menjaga precision
# ============================================================

import optuna
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_predict
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    precision_score,
    recall_score,
    fbeta_score,
    f1_score,
    accuracy_score,
    balanced_accuracy_score,
    roc_auc_score,
    average_precision_score,
    classification_report,
    confusion_matrix
)

target = 'is_default'

# ============================================================
# 1. Fitur Random Forest
# ============================================================

features_rf = [
    'usia',
    'jumlah_tanggunga',
    'pendapatan_per_b',
    'lama_bekerja_tah',
    'jumlah_kredit_ak',
    'total_kredit',
    'sisa_bayar_kredi',
    'tenor_bulan',
    'bunga_per_tahun',
    'cicilan_per_bula',
    'rasio_utang_pend',
    'jumlah_tunggakan',
    'skor_kredit_sebe',
    'jenis_kelamin',
    'status_pernikaha',
    'jenis_pekerjaan',
    'pendidikan',
    'kota',
    'jenis_kredit'
]

features_rf = [col for col in features_rf if col in df.columns]

df_rf = df[features_rf + [target]].dropna().copy()

X_rf = df_rf[features_rf]
y_rf = df_rf[target].astype(int)

X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
    X_rf,
    y_rf,
    test_size=0.20,
    random_state=42,
    stratify=y_rf
)

numeric_features_rf = X_rf.select_dtypes(include=[np.number]).columns.tolist()
categorical_features_rf = X_rf.select_dtypes(include=['object']).columns.tolist()

preprocessor_rf = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numeric_features_rf),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features_rf)
    ]
)

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

```python
# ============================================================
# 2. Objective function Optuna
# ============================================================

def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 300, 1000),
        'max_depth': trial.suggest_int('max_depth', 4, 25),
        'min_samples_split': trial.suggest_int('min_samples_split', 5, 80),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 2, 40),
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', 0.3, 0.5, 0.7]),
        'bootstrap': trial.suggest_categorical('bootstrap', [True, False]),
        'class_weight': trial.suggest_categorical(
            'class_weight',
            [
                'balanced',
                'balanced_subsample',
                {0: 1, 1: 2},
                {0: 1, 1: 3},
                {0: 1, 1: 4},
                {0: 1, 1: 5},
                {0: 1, 1: 8}
            ]
        ),
        'random_state': 42,
        'n_jobs': -1
    }
    
    model = RandomForestClassifier(**params)
    
    pipeline = Pipeline(
        steps=[
            ('preprocessor', preprocessor_rf),
            ('model', model)
        ]
    )
    
    # Cross-val predicted probability
    y_proba_cv = cross_val_predict(
        pipeline,
        X_train_rf,
        y_train_rf,
        cv=cv,
        method='predict_proba',
        n_jobs=-1
    )[:, 1]
    
    # Threshold ikut dituning
    threshold = trial.suggest_float('threshold', 0.15, 0.60)
    y_pred_cv = (y_proba_cv >= threshold).astype(int)
    
    precision = precision_score(y_train_rf, y_pred_cv, zero_division=0)
    recall = recall_score(y_train_rf, y_pred_cv, zero_division=0)
    f2 = fbeta_score(y_train_rf, y_pred_cv, beta=2, zero_division=0)
    
    # Penalti kalau precision terlalu rendah
    # Biar model tidak asal menebak default semua
    if precision < 0.35:
        f2 = f2 * 0.7
    
    return f2
```

```python
# ============================================================
# 3. Run Optuna Study
# ============================================================

study = optuna.create_study(direction='maximize')

study.optimize(
    objective,
    n_trials=50,
    show_progress_bar=True
)

print("Best F2 Score:")
print(study.best_value)

print("\nBest Parameters:")
print(study.best_params)
```

```python
# ============================================================
# OPTUNA RANDOM FOREST - 3 OBJECTIVES
# F1, PRECISION, ACCURACY
# ============================================================

import optuna
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score,
    accuracy_score,
    balanced_accuracy_score,
    roc_auc_score,
    average_precision_score,
    classification_report,
    confusion_matrix
)
```

```python
# ============================================================
# FUNCTION: Mapping class_weight
# ============================================================

class_weight_map = {
    'balanced': 'balanced',
    'balanced_subsample': 'balanced_subsample',
    'custom_2': {0: 1, 1: 2},
    'custom_3': {0: 1, 1: 3},
    'custom_4': {0: 1, 1: 4},
    'custom_5': {0: 1, 1: 5}
}


# ============================================================
# FUNCTION: Objective Optuna
# ============================================================

def create_objective(metric_name):
    def objective(trial):
        class_weight_option = trial.suggest_categorical(
            'class_weight_option',
            [
                'balanced',
                'balanced_subsample',
                'custom_2',
                'custom_3',
                'custom_4',
                'custom_5'
            ]
        )
        
        class_weight = class_weight_map[class_weight_option]
        
        params = {
            'n_estimators': trial.suggest_int('n_estimators', 300, 900),
            'max_depth': trial.suggest_int('max_depth', 4, 25),
            'min_samples_split': trial.suggest_int('min_samples_split', 5, 80),
            'min_samples_leaf': trial.suggest_int('min_samples_leaf', 2, 40),
            'max_features': trial.suggest_categorical(
                'max_features',
                ['sqrt', 'log2', 0.3, 0.5, 0.7]
            ),
            'bootstrap': trial.suggest_categorical('bootstrap', [True, False]),
            'class_weight': class_weight,
            'random_state': 42,
            'n_jobs': -1
        }
        
        model = RandomForestClassifier(**params)
        
        pipeline = Pipeline(
            steps=[
                ('preprocessor', preprocessor_rf),
                ('model', model)
            ]
        )
        
        # Cross-validation probability prediction
        y_proba_cv = cross_val_predict(
            pipeline,
            X_train_rf,
            y_train_rf,
            cv=cv,
            method='predict_proba',
            n_jobs=-1
        )[:, 1]
        
        # Threshold ikut dituning
        threshold = trial.suggest_float('threshold', 0.20, 0.75)
        y_pred_cv = (y_proba_cv >= threshold).astype(int)
        
        precision = precision_score(y_train_rf, y_pred_cv, zero_division=0)
        recall = recall_score(y_train_rf, y_pred_cv, zero_division=0)
        f1 = f1_score(y_train_rf, y_pred_cv, zero_division=0)
        accuracy = accuracy_score(y_train_rf, y_pred_cv)
        
        # Objective sesuai metric
        if metric_name == 'f1':
            score = f1
            
            # Penalti jika precision dan recall terlalu timpang
            gap = abs(precision - recall)
            if gap > 0.30:
                score = score * 0.90
        
        elif metric_name == 'precision':
            score = precision
            
            # Penalti jika recall terlalu rendah
            # Supaya model tidak terlalu konservatif memprediksi default
            if recall < 0.50:
                score = score * 0.75
        
        elif metric_name == 'accuracy':
            score = accuracy
            
            # Penalti jika recall default terlalu rendah
            # Karena accuracy bisa bias ke non-default
            if recall < 0.50:
                score = score * 0.80
        
        else:
            raise ValueError("metric_name harus 'f1', 'precision', atau 'accuracy'")
        
        return score
    
    return objective
```

```python
# ============================================================
# RUN 3 OPTUNA STUDIES
# ============================================================

N_TRIALS = 50

study_f1 = optuna.create_study(direction='maximize')
study_f1.optimize(
    create_objective('f1'),
    n_trials=N_TRIALS,
    show_progress_bar=True
)

study_precision = optuna.create_study(direction='maximize')
study_precision.optimize(
    create_objective('precision'),
    n_trials=N_TRIALS,
    show_progress_bar=True
)

study_accuracy = optuna.create_study(direction='maximize')
study_accuracy.optimize(
    create_objective('accuracy'),
    n_trials=N_TRIALS,
    show_progress_bar=True
)

print("Best F1 study:")
print(study_f1.best_value)
print(study_f1.best_params)

print("\nBest Precision study:")
print(study_precision.best_value)
print(study_precision.best_params)

print("\nBest Accuracy study:")
print(study_accuracy.best_value)
print(study_accuracy.best_params)
```

## Slide 28 K-Means

```python
# ============================================================
# SLIDE 28 - SEGMENTASI NASABAH DENGAN K-MEANS CLUSTERING
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

target = 'is_default'

cluster_features = [
    'jenis_pekerjaan',
    'pendapatan_per_b',
    'rasio_utang_pend',
    'jumlah_tunggakan',
    'usia',
    'skor_kredit_sebe'
]

cluster_features = [col for col in cluster_features if col in df.columns]

df_cluster = df[cluster_features + [target]].dropna().copy()

X_cluster = df_cluster[cluster_features]

num_cols_cluster = X_cluster.select_dtypes(include=[np.number]).columns.tolist()
cat_cols_cluster = X_cluster.select_dtypes(include=['object', 'category']).columns.tolist()

preprocessor_cluster = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols_cluster),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols_cluster)
    ]
)

X_cluster_processed = preprocessor_cluster.fit_transform(X_cluster)

# ============================================================
# 1. Tentukan jumlah cluster: Elbow Method + Silhouette Score
# ============================================================

k_range = range(2, 11)

inertia_scores = []
silhouette_scores = []

for k in k_range:
    kmeans_temp = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    
    labels_temp = kmeans_temp.fit_predict(X_cluster_processed)
    
    inertia_scores.append(kmeans_temp.inertia_)
    silhouette_scores.append(silhouette_score(X_cluster_processed, labels_temp))

cluster_eval = pd.DataFrame({
    'k': list(k_range),
    'inertia': inertia_scores,
    'silhouette_score': silhouette_scores
})

display(cluster_eval)

# Plot elbow
plt.figure(figsize=(7, 4))
plt.plot(cluster_eval['k'], cluster_eval['inertia'], marker='o')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method - KMeans')
plt.grid(True)
plt.show()

# Plot silhouette
plt.figure(figsize=(7, 4))
plt.plot(cluster_eval['k'], cluster_eval['silhouette_score'], marker='o')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score - KMeans')
plt.grid(True)
plt.show()

# Pilih k terbaik berdasarkan silhouette tertinggi
best_k = cluster_eval.loc[cluster_eval['silhouette_score'].idxmax(), 'k']
print("Jumlah cluster terpilih berdasarkan silhouette score:", best_k)

# ============================================================
# 2. Final KMeans
# ============================================================

kmeans_final = KMeans(
    n_clusters=int(best_k),
    random_state=42,
    n_init=10
)

df_cluster['cluster'] = kmeans_final.fit_predict(X_cluster_processed)

# ============================================================
# 3. Profil cluster + default rate
# target hanya dipakai untuk evaluasi profil, bukan untuk clustering
# ============================================================

cluster_profile_num = df_cluster.groupby('cluster').agg(
    jumlah_nasabah=(target, 'count'),
    jumlah_default=(target, 'sum'),
    default_rate=(target, 'mean'),
    usia_rata2=('usia', 'mean'),
    pendapatan_rata2=('pendapatan_per_b', 'mean'),
    dti_rata2=('rasio_utang_pend', 'mean'),
    tunggakan_rata2=('jumlah_tunggakan', 'mean'),
    skor_kredit_rata2=('skor_kredit_sebe', 'mean')
).reset_index()

cluster_profile_num['persen_nasabah'] = (
    cluster_profile_num['jumlah_nasabah'] / len(df_cluster) * 100
).round(2)

cluster_profile_num['default_rate'] = (
    cluster_profile_num['default_rate'] * 100
).round(2)

cluster_profile_num = cluster_profile_num.sort_values(
    by='default_rate',
    ascending=False
)

display(cluster_profile_num)

# Ciri khas kategorik: modus jenis pekerjaan
cluster_profile_cat = df_cluster.groupby('cluster').agg(
    pekerjaan_dominan=('jenis_pekerjaan', lambda x: x.mode()[0])
).reset_index()

cluster_profile = cluster_profile_num.merge(
    cluster_profile_cat,
    on='cluster',
    how='left'
)

display(cluster_profile)

# Cluster dengan default rate tertinggi
highest_risk_cluster = cluster_profile.iloc[0]

print("Cluster dengan tingkat gagal bayar tertinggi:")
display(highest_risk_cluster.to_frame().T)
```
